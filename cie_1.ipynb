```python
import pandas as pd
import matplotlib.pyplot as plt
iris=pd.read_csv('/content/flowers.csv')
x=iris['petal length']
y=iris['petal width']
species_counts=iris['species'].value_counts()
plt.bar(species_counts.index,species_counts.values,color="lightblue")
plt.xlabel('species')
plt.ylabel('count')
plt.title('petal length')
plt.grid(axis='y')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
iris=pd.read_csv('/content/flowers.csv')
x=iris['petal length']
y=iris['petal width']
plt.hist(iris['petal length'],bins=10,color='green',edgecolor='black')
plt.title("Distribution of Petal Length")
plt.xlabel("Petal Length")
plt.ylabel("petal width")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
iris=pd.read_csv('/content/flowers.csv')
x=iris['petal length']
y=iris['petal width']
species = iris['species'].unique()
data = [iris[iris['species'] == s]['petal length'] for s in species]
plt.boxplot('species'['petal length'])
plt.title("Petal Length by Species")
plt.xlabel("Species")
plt.ylabel("Petal Length")
plt.grid(True)
plt.show()
```
